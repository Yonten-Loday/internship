```python
import pandas as pd
import numpy as np
```

```python
#sql query over pandas dataframe
from pandasql import sqldf
import pandasql as ps
#!pip install sqldf
#!pip install pandasql

secondDraft = pd.read_csv('secondDraft.csv',index_col = 0)
secondDraft = ps.sqldf("SELECT * FROM secondDraft")
secondDraft
```

```python
df_bank = secondDraft
if len(df_bank['bank_account']) == 9:
    if (ps.sqldf("SELECT * FROM df_bank WHERE (bank_account LIKE '2%' or bank_account LIKE '1%') and bank_account_type = 'Saving Account'")):
        df_bank['bank_name'] = 'BANK OF BHUTAN LIMITED'
        
elif len(df_bank['bank_account']) == 12:
    if (ps.sqldf("SELECT * FROM df_bank WHERE (bank_account LIKE '00%') and bank_account_type = 'Saving Account'")):
        df_bank['bank_name'] = 'BHUTAN DEVELOPMENT BANK LIMITED'
    elif (ps.sqldf("SELECT * FROM df_bank WHERE bank_account LIKE '11%' and bank_account_type = 'Saving Account'")):
        df_bank['bank_name'] = 'DRUK PNB BANK LIMITED'
    elif (ps.sqldf("SELECT * FROM df_bank WHERE bank_account LIKE '10%' and bank_account_type = 'Saving Account'")):
        df_bank['bank_name'] = 'DIGITAL KIDU BANK'

elif len(df_bank['bank_account']) == 13: 
    if (ps.sqldf("SELECT * FROM df_bank WHERE bank_account LIKE '51%' and bank_account_type = 'Saving Account'")):
        df_bank['bank_name'] = 'BHUTAN NATIONAL BANK LIMITED'

elif len(df_bank['bank_account']) == 14:
    if (ps.sqldf("SELECT * FROM df_bank WHERE bank_account LIKE '77%' and bank_account_type = 'Saving Account'")):
        df_bank['bank_name'] = 'T BANK LIMITED'
        
elif len(df_bank['bank_account']) <= 8:
    df_bank['bank_name'] = 'Unknown'
        
df_bank     
```
